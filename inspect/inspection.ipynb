# Inspection

```python
import polars as pl
```

```python
df = pl.read_csv("realtor-data.csv")
```

```python
print(df.schema)  # column names and inferred types
```

```python
print(df.shape)  # (row_count, col_count)
```

```python
print(df.null_count())  # nulls per column
```

```python
print(df.head(5))
```

```python
df = df.filter(
    pl.col("price").is_not_null()
    | pl.col("state").is_not_null()
    | pl.col("zip_code").is_not_null()
)
print(df.shape)
```
